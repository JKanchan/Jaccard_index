```python
import pandas as pd
import numpy as np
```

```python
col_name = ['Company Name']
```

```python
jaccard_index = np.zeros((100,100))
```

```python
df = pd.read_csv('data.csv', skipinitialspace=True, usecols=col_name)
```

```python
df['Company Name'] = df['Company Name'].str.split(' ')
```

```python
df['Company Name'] = df['Company Name'].apply(set)
```

```python
df['Company Name'].loc[24]
```

```python
col = df['Company Name'].values
```

```python
def jaccard_similarity_score(a, b):
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
for i, set1 in enumerate(col):
    for j, set2 in enumerate(col[1:]):
        jaccard_index[i][j] =jaccard_similarity_score(set1,set2)
```

```python
list_of_rows = []
```

```python
for i in range(len(col)):
    list1 = []
    for j in range(len(col)):
        if(jaccard_index[i][j]>0.5):
            print(df['Company Name'].loc[j])
            list1.append(str(j))
    list_of_rows.append(list1)       
```

```python
df1 = pd.DataFrame(list_of_rows)
```

```python
df1 = df1.drop_duplicates()
```
